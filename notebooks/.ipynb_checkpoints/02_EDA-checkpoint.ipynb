```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../data/german_credit_data.csv")
df.head()
```

```python
sns.countplot(x='Risk',data=df)
plt.title("Risk Distribution")
plt.show()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
sns.histplot(x='Age',bins=20,data=df)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(x='Credit amount',bins=30,data=df)
plt.title("Credit Amount")
plt.show()
```

```python
sns.histplot(x='Duration',bins=20,data=df)
plt.title("Duration")
plt.show()
```

```python
df['Sex'].value_counts()
```

```python
sns.countplot(x='Sex',data=df)
plt.title("Sex")
plt.show()
```

```python
df["Housing"].value_counts()
```

```python
sns.countplot(x='Housing',data=df)
plt.title("Housing")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y="Purpose",data=df)
plt.show()
```

```python
sns.boxplot(x='Age',data=df)
plt.show()
```
